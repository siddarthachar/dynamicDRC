Oxidation of $CO$ to form $O_2$ using $Pt(111)$ catalyst at 300 K. <br>
The pressure for $CO$ is 10 bar and the pressure for $O_2$ is 1 bar. The partial pressure of $CO_2$ is considered to be 0. 

Overall reaction: <br>
$CO + O_2 \rightarrow CO_2$

The reaction proceeds in the following way:  <br>
R1:$O_2 + 2* \rightarrow 2O*$ <br>
R2:$CO + * \rightarrow CO*$ <br>
R3:$O* + CO* \rightarrow CO_2 + 2*$<br>

Energies (all in eV) obtained from CATAPP: <br>
$G_{TS_1} = 0$ <br>
$G_{TS_2} = -0.62$ <br>
$G_{TS_3} = -0.72$ <br>
$G_{O*} = -1.5/2 = -0.72$ <br>
$G_{CO*} = -1.2$ <br>
$G_{CO_2} = -0.96$ <br>

The mole balances used are: <br> 
$\frac{\theta_{O*}}{dt} = k_1P_{O_2}{\theta_{*}}^2 - k_{-1}{\theta_{O*}}^2 - k_{3}\theta_{O*}\theta_{CO*} + k_{-3}P_{CO_2}{\theta_{*}}^2$

$\frac{\theta_{CO*}}{dt} = k_2P_{CO*}\theta_{*} - k_{-2}\theta_{CO*} - k_{3}\theta_{O*}\theta_{CO*} + k_{-3}P_{CO_2}{\theta_{*}}^2$
